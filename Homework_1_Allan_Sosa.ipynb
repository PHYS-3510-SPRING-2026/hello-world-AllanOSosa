```python
# Problem 1 - Gauss' Sum

# Sum by brute force
A_1 = 1+2+3+4+5+6+7
A_2 = 10+11+12+13+14+15+16
A_3 = 20+21+22+23+24+25+26

# Sum by Gauss' Formula
n = 7
F = 1
L = 7

# Plug in all the variables to Gauss' Formula
S = n * (F+L)/2

# Determine whether the two methods produce the same result
C = (A_1==S)

# final Answer
print("The statement", A_1, "and", S, "are equivalent is", C)
```

```python
# Problem 2 - Gravitational Force

# Variables for Newton's Law of Gravitation
G = 6.6743*(10**-11)
m1 = 1.0e25
m2 = 1.2e22
r = 4.1e7

# Computing the Gravitational Force by using the given variables
F = G*(m1*m2)/(r**2)

# Final Answer
print("The magnitude of the Gravitation Force is", F, "N" )
```

```python
# Problem 3 - Congruence Modulo 17
A = 1234
B = 2586
C = 17

# Compute the Modulo Operation
A_1 = A % C
A_2 = B % C

# Compute the Equality Operation
E = (A_1==A_2)

# Final Answer
if(E==True):
  print("The statement for", A_1, "and", A_2, "shows that both integers are congruent modulo 17")
if(E==False):
  print("The statement for", A_1, "and", A_2, "shows that both integers are not congruent modulo 17")

```
